```python
#3 Для каждого атрибута нужно определить:
#a.среднее значение, ско
#b.построить гистограмму распределения значений, определить есть ли выбросы 
#c.есть ли пропущенные значение, сколько
#d.предложить вариант обработки пропущенных значений

#a:
df = pd.DataFrame.from_dict(mydata)
del df['Substitution']
df[df.columns.intersection(mydata)]
print('a. Average value\n--------------------------')
print(df.loc[:, ['Matches_Played', 'Mins', 'Goals', 'xG', 'xG Per Avg Match', 'Shots', 'OnTarget',
                 'Shots Per Avg Match', 'On Target Per Avg Match']].mean(),'\n\n') 

print('std\n--------------------------')
print(df.loc[:, ['Matches_Played', 'Mins', 'Goals', 'xG', 'xG Per Avg Match', 'Shots', 'OnTarget', 
                 'Shots Per Avg Match', 'On Target Per Avg Match']].std(), '\n\n') 

#b(построим гистограмму распределения значений для каждого числового атрибута):
print('Histogram\n')
for x in ['Matches_Played', 'Mins', 'Goals', 'xG', 'xG Per Avg Match', 'Shots', 'OnTarget', 
          'Shots Per Avg Match', 'On Target Per Avg Match']:
    plt.figure(figsize=(15, 5))
    plt.hist(mydata[x], color='green', bins=18, stacked=False, edgecolor='black')
    plt.xlabel(x)
    plt.ylabel('Amount')
    plt.show()
    
# #Осуществим поиск выбросов путем построения коробчатых диаграм
for x in ['Matches_Played', 'Mins', 'Goals', 'xG', 'xG Per Avg Match', 'Shots', 
          'OnTarget', 'Shots Per Avg Match', 'On Target Per Avg Match']:
    plt.boxplot(mydata[x])
    plt.show()
    
#c. Найдем количество пропущенных значений в каждой колонке с помощью метода isnull:
print('\n\nMissing values\n-------------------------')
df.isnull().sum()
```

```python

#d. Количество пропуссков - 34, все они в атрибуте club, соответственно, данные пропуски не влияюют на числовые атрибуты.
#В текущей ситуации пропуски являются ошибками, которые можно устранить путем дополнительного сбора данных из других источников.
#4. a
print('\n\nCorrelation matrix\n-------------------------')
x = ['Matches_Played', 'Mins', 'Goals', 'xG', 'xG Per Avg Match', 'Shots', 
     'OnTarget', 'Shots Per Avg Match', 'On Target Per Avg Match']
corr = mydata[x].corr()
corr.style.background_gradient(cmap='RdYlGn')


#b. Корреляция отсутствует у факторов не влияющих на обрабатываемые данные, а именно: Country, League, Club, Player Names. 
#Также корреляция близка к нулю у следующих атрибутов: Matches_Played - Shots Per Avg Match; 
#Matches_Played - On Target Per Avg Match; Mins - Shots Per Avg Match; Mins - On Target Per Avg Match
```

```python
#c. Построим матрицу графиков рассеивания с графиками оценки плотности ядра по диагонали
pd.plotting.scatter_matrix(mydata[x], diagonal = 'kde')
plt.show() 


```
